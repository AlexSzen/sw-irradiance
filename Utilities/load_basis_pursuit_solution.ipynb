```python
#Note the image size here (512,512) and the lgtaxis are different than what we'be been using so far!

def AIA_load(year,month,day):
    import numpy as np
    dim = [512,512]
    nchannels = 6
    nlogT = 18
    lgtaxis = np.arange(nlogT)*0.1 + 5.45
    
    path = '/data/NASAFDL2018/SpaceWeather/Team1-Meng/SunInTime/'
    path = path + 'AIA_DEM_{0:04d}-{1:02d}-{2:02d}'.format(year, month, day)
    
    fp = np.memmap(path+'.aia',mode='r', shape=(nchannels,dim[1],dim[0]),dtype='float64')
    img = np.zeros((dim[0],dim[1],nchannels))
    img[:] = np.transpose(fp[:],(2,1,0))
    del fp
    
    fp = np.memmap(path+'.emcube',mode='r', shape=(nlogT,dim[1],dim[0]),dtype='float32')
    emcube = np.zeros((dim[0],dim[1],nlogT))
    emcube[:] = np.transpose(fp[:],(2,1,0))
    del fp
    
    return img, emcube, lgtaxis
```

```python
img, emcube, lgtaxis = AIA_load(2011,4,2)
```
