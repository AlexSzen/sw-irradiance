```python
import pdb
import numpy as np
import PIL.Image
import os
import sunpy as sp
import matplotlib.pyplot as plt
%matplotlib inline

from IPython.core.display import display, HTML
display(HTML('<style>.container { width:98% !important; }</style>'))
```

```python
# loading data 
X = np.load("/scratch/df_spatial_visualizations/anet_2_0/anet_2_0.npy")
Xi = np.load("/scratch/df_spatial_visualizations/anet_2_0/anet_2_0.npy.input.npy")
```

```python
# pick a random image in the dump
im_i = (np.random.rand(1)[0]*X.shape[0]).astype(int)

AIA = Xi[im_i,:,:,:]
EVE = X[im_i,:,:,:]
AIA.shape
```

##  AIA + EVE

```python


# Size definitions
dpi = 200
pxx = 750   # Horizontal size of each panel
pxy = 750   # Vertical size of each panel

nph = 6      # Number of horizontal panels
npv = 2      # Number of vertical panels
frc = 1    # Fraction of the panel devoted to histograms

# Padding
padv  = 100 #Vertical padding in pixels
padv2 = 0  #Vertical padding in pixels between panels
padh  = 100 #Horizontal padding in pixels at the edge of the figure
padh2 = 0 #Horizontal padding in pixels between panels

# Figure sizes in pixels
fszv = (npv*pxy + 2*padv + (npv-1)*padv2 )      #Vertical size of figure in pixels
fszh = (nph*pxx + 2*padh + (nph-1)*padh2 )      #Horizontal size of figure in pixels

# Conversion to relative unites
ppadv  = padv/fszv     #Vertical padding in relative units
ppadv2 = padv2/fszv    #Vertical padding in relative units

ppadh  = padh/fszh     #Horizontal padding the edge of the figure in relative units
ppadh2 = padh2/fszh    #Horizontal padding between panels in relative units

fig = plt.figure(figsize=(fszh/dpi,fszv/dpi))

ty = 0.95

# AIA 94
ax1 = fig.add_axes([ppadh, ppadv + pxy/fszv, pxx/fszh, pxy/fszv])
ax1.pcolor(AIA[0,:,:],vmin=np.min(AIA[0,:,:]), vmax=np.max(AIA[0,:,:]), cmap = plt.get_cmap('sdoaia94'))
ax1.text(0.5,ty, '94Å', color = 'w', fontsize = 20 , horizontalalignment='center', verticalalignment='center', transform=ax1.transAxes)
ax1.axis('off')

ax1b = fig.add_axes([ppadh, ppadv, pxx/fszh, pxy/fszv])
ax1b.pcolor(EVE[0,:,:],vmin=np.min(EVE[0,:,:]), vmax=np.max(EVE[0,:,:]), cmap = plt.get_cmap('sdoaia94'))
ax1b.text(0.5,ty, 'Fe XVIII', color = 'w', fontsize = 20 , horizontalalignment='center', verticalalignment='center', transform=ax1b.transAxes)
ax1b.axis('off')




# AIA 131
ax2 = fig.add_axes([ppadh + pxx/fszh , ppadv + pxy/fszv, pxx/fszh, pxy/fszv])
ax2.pcolor(AIA[1,:,:],vmin=np.min(AIA[1,:,:]), vmax=np.max(AIA[1,:,:]), cmap = plt.get_cmap('sdoaia131'))
ax2.text(0.5,ty, '131Å', color = 'w', fontsize = 20 , horizontalalignment='center', verticalalignment='center', transform=ax2.transAxes)
ax2.axis('off')

ax2b = fig.add_axes([ppadh + pxx/fszh , ppadv, pxx/fszh, pxy/fszv])
ax2b.pcolor(EVE[1,:,:]+EVE[2,:,:],vmin=np.min(EVE[1,:,:]+EVE[2,:,:]), vmax=np.max(EVE[1,:,:]+EVE[2,:,:]), cmap = plt.get_cmap('sdoaia131'))
ax2b.text(0.5,ty, 'Fe VIII + Fe XX', color = 'w', fontsize = 20 , horizontalalignment='center', verticalalignment='center', transform=ax2b.transAxes)
ax2b.axis('off')



# AIA 171
ax3 = fig.add_axes([ppadh + 2*pxx/fszh , ppadv + pxy/fszv, pxx/fszh, pxy/fszv])
ax3.pcolor(AIA[2,:,:],vmin=np.min(AIA[2,:,:]), vmax=np.max(AIA[2,:,:]), cmap = plt.get_cmap('sdoaia171'))
ax3.text(0.5,ty, '171Å', color = 'w', fontsize = 20 , horizontalalignment='center', verticalalignment='center', transform=ax3.transAxes)
ax3.axis('off')

ax3b = fig.add_axes([ppadh + 2*pxx/fszh , ppadv, pxx/fszh, pxy/fszv])
ax3b.pcolor(EVE[3,:,:],vmin=np.min(EVE[3,:,:]), vmax=np.max(EVE[3,:,:]), cmap = plt.get_cmap('sdoaia171'))
ax3b.text(0.5,ty, 'Fe IX', color = 'w', fontsize = 20 , horizontalalignment='center', verticalalignment='center', transform=ax3b.transAxes)
ax3b.axis('off')







# AIA 193
ax4 = fig.add_axes([ppadh + 3*pxx/fszh , ppadv + pxy/fszv, pxx/fszh, pxy/fszv])
ax4.pcolor(AIA[3,:,:],vmin=np.min(AIA[3,:,:]), vmax=np.max(AIA[3,:,:]), cmap = plt.get_cmap('sdoaia193'))
ax4.text(0.5,ty, '193Å', color = 'w', fontsize = 20 , horizontalalignment='center', verticalalignment='center', transform=ax4.transAxes)
ax4.axis('off')

ax4b = fig.add_axes([ppadh + 3*pxx/fszh , ppadv, pxx/fszh, pxy/fszv])
ax4b.pcolor(EVE[6,:,:]+EVE[7,:,:],vmin=np.min(EVE[6,:,:]+EVE[7,:,:]), vmax=np.max(EVE[6,:,:]+EVE[7,:,:]), cmap = plt.get_cmap('sdoaia193'))
ax4b.text(0.5,ty, 'Fe XII + Fe XIII', color = 'w', fontsize = 20 , horizontalalignment='center', verticalalignment='center', transform=ax4b.transAxes)
ax4b.axis('off')





# AIA 211
ax5 = fig.add_axes([ppadh + 4*pxx/fszh , ppadv + pxy/fszv, pxx/fszh, pxy/fszv])
ax5.pcolor(AIA[4,:,:],vmin=np.min(AIA[4,:,:]), vmax=np.max(AIA[4,:,:]), cmap = plt.get_cmap('sdoaia211'))
ax5.text(0.5,ty, '211Å', color = 'w', fontsize = 20 , horizontalalignment='center', verticalalignment='center', transform=ax5.transAxes)
ax5.axis('off')

ax5b = fig.add_axes([ppadh + 4*pxx/fszh , ppadv, pxx/fszh, pxy/fszv])
ax5b.pcolor(EVE[8,:,:],vmin=np.min(EVE[8,:,:]), vmax=np.max(EVE[8,:,:]), cmap = plt.get_cmap('sdoaia211'))
ax5b.text(0.5,ty, '211Å', color = 'w', fontsize = 20 , horizontalalignment='center', verticalalignment='center', transform=ax5b.transAxes)
ax5b.axis('off')




# AIA 335 Flipped with 304
ax6 = fig.add_axes([ppadh + 6*pxx/fszh , ppadv + pxy/fszv, pxx/fszh, pxy/fszv])
ax6.pcolor(AIA[5,:,:],vmin=np.min(AIA[5,:,:]), vmax=np.max(AIA[5,:,:]), cmap = plt.get_cmap('sdoaia335'))
ax6.text(0.5,ty, '335Å', color = 'w', fontsize = 20 , horizontalalignment='center', verticalalignment='center', transform=ax6.transAxes)
ax6.axis('off')

ax6b = fig.add_axes([ppadh + 6*pxx/fszh , ppadv, pxx/fszh, pxy/fszv])
ax6b.pcolor(EVE[12,:,:],vmin=np.min(EVE[12,:,:]), vmax=np.max(EVE[12,:,:]), cmap = plt.get_cmap('sdoaia335'))
ax6b.text(0.5,ty, 'Fe XVI', color = 'w', fontsize = 20 , horizontalalignment='center', verticalalignment='center', transform=ax6b.transAxes)
ax6b.axis('off')



# AIA 304 Flipped with 335
ax7 = fig.add_axes([ppadh + 5*pxx/fszh , ppadv + pxy/fszv, pxx/fszh, pxy/fszv])
ax7.pcolor(AIA[6,:,:],vmin=np.min(AIA[6,:,:]), vmax=np.max(AIA[6,:,:]), cmap = plt.get_cmap('sdoaia304'))
ax7.text(0.5,ty, '304Å', color = 'w', fontsize = 20 , horizontalalignment='center', verticalalignment='center', transform=ax7.transAxes)
ax7.axis('off')

ax7b = fig.add_axes([ppadh + 5*pxx/fszh , ppadv, pxx/fszh, pxy/fszv])
ax7b.pcolor(EVE[11,:,:],vmin=np.min(EVE[11,:,:]), vmax=np.max(EVE[11,:,:]), cmap = plt.get_cmap('sdoaia304'))
ax7b.text(0.5,ty, 'He II', color = 'w', fontsize = 20 , horizontalalignment='center', verticalalignment='center', transform=ax7b.transAxes)
ax7b.axis('off')
```
